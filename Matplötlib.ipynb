```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
#Scatter Pl√∂t

x=np.random.rand(1000)
y=np.random.rand(1000)

plt.scatter(x,y,color='green',marker='*', alpha=0.3)
```

```python
#Line Pl√∂t

years = [2000 + x for x in range(10)]
salary = [20,30,40,10,90,40,20,60,70,20,]

print(years)
print(salary)

plt.plot(years,salary, c='red',lw=5,linestyle='--')
```

```python
#Bar Chart

titles = [4,0,5,1,0,1,3,9]
ipl_team = ["CSK","RCB","MI","GT","PSK","DC","RR","LSG"]

print(titles)
print(ipl_team)

plt.bar(ipl_team,titles, color='red',align='edge')
```

```python
#Histogram

ages = np.random.normal(6,2,100)

plt.hist(ages, bins=50,color='g',cumulative=True)
```

```python
titles = [30,90,50,20,10]
ipl_team = ["cycle","bike","train","flight","zet"]
explodes = [0,0,0,0,1]
style.use("ggplot")

plt.pie(titles, labels=ipl_team,explode=explodes, autopct="%.2f%%",pctdistance=0.6, startangle=180)
plt.legend(title="transport")
```

```python
titles = np.random.random(100)
plt.boxplot(titles)
```

```python
#Line Pl√∂t

years = [2000 + x for x in range(10)]
runs = [100,120,130,150,170,180,200,230,250,280]

style.use("ggplot")

plt.plot(years,runs, c='blue')
plt.title("No. of Runs in IPL increasing every year")
plt.xlabel("Year",fontsize=10)
plt.ylabel("Runs")
plt.legend()
```

```python
#SUB Plotting

x=np.random.rand(1000)

fig, axs = plt.subplots(2,2)

axs[0,0].plot(x,np.sin(x))
axs[0,0].set_title("Log_Graph")

axs[0,1].plot(x,np.cos(x))
axs[0,1].set_title("Log_Graph")

axs[1,0].plot(x,np.tan(x))
axs[1,0].set_title("Log_Graph")

axs[1,1].plot(x,np.log(x))
axs[1,1].set_title("Log_Graph")

plt.savefig("4plot.png", dpi=400)
```

```python
#3D plots


x = np.arange(1,10,0.5)
y = np.arange(10,20,0.5)

X,Y = np.meshgrid(x,y)

Z=np.sin(X)*np.cos(Y)

ax=plt.axes(projection="3d")

#ax.scatter(x,y,z)

ax.plot_surface(X,Y,Z)
ax.set_title("3D Plot")
ax.set_xlabel("X-Axis")
ax.set_ylabel("Y-Axis")
ax.set_zlabel("Z-Axis")
```
